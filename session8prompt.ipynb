```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sine(x):
    return np.sin(x)

def cosine(x):
    return np.cos(x)
```

```python
x = np.linspace(0, 1, 100)
y = sine(x)
z = cosine(x)
```

```python
def plot(x, y,
         flag_save = True,
         fname='session8prompt.pdf'):
    f, ax = plt.subplots(1,2)

    for i in range(1):
        ax[i].plot(x, y, color = "red")
        ax[i].set_xlabel("x")
        ax[i].set_ylabel("sin(x)")

    for j in range(2):
        if j == 1:
            ax[j].plot(x, z, color = "blue")
            ax[j].set_xlabel("x")
            ax[j].set_ylabel("cos(x)")

    plt.tight_layout()

    if(flag_save): 
            plt.savefig(fname)


def main(): 
    x, y, z 
    plot(x,y)


if __name__ == '__main__':
    main()
```
